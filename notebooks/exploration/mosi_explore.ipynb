```python
def cmumosei_round(a):
    if a < -2:
        res = -3
    if -2 <= a and a < -1:
        res = -2
    if -1 <= a and a < 0:
        res = -1
    if 0 <= a and a <= 0:
            res = 0
    if 0 < a and a <= 1:
            res = 1
    if 1 < a and a <= 2:
            res = 2
    if a > 2:
            res = 3
    return res

print(data['train']['labels'])
labels = []
for i in data['train']['labels']:
      if type(i[0]).__name__ == 'ndarray':
        labels.append(i[0][0])
        continue
      else:
        labels.append(i[0])
print(labels)
print(len(labels))

labelset = sorted(list(set([cmumosei_round(label) for label in labels])))
print(labelset)
```
